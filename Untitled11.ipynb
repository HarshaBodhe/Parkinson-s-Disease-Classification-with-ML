```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.neural_network import MLPClassifier
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Conv1D, MaxPooling1D, Flatten, Dropout, Reshape
from tensorflow.keras.optimizers import Adam
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("pd_speech_features.csv", header=[1])
df = df.drop(index=0).reset_index(drop=True)
df_clean = df.drop(columns=['id'])
```

```python
non_feature_cols = ['gender', 'class']
for col in df_clean.columns:
    if col not in non_feature_cols:
        df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')

df_clean = df_clean.fillna(df_clean.mean())
```

```python
print("Class Distribution:")
print(df_clean['class'].value_counts())
sns.countplot(x='class', data=df_clean)
plt.title('Class Distribution')
plt.show()
corr = df_clean.drop(columns=['gender']).corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr[['class']].sort_values(by='class', ascending=False), annot=True)
plt.title('Correlation with Class')
plt.show()
```

```python
X = df_clean.drop(columns=['class'])
y = df_clean['class'].astype(int)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)
```

```python
log_reg = LogisticRegression(max_iter=1000, random_state=42)
log_reg.fit(X_train, y_train)
y_pred_lr = log_reg.predict(X_test)
print("\nLogistic Regression")
print("Accuracy:", accuracy_score(y_test, y_pred_lr))
print(classification_report(y_test, y_pred_lr))
```

```python
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
print("\nRandom Forest")
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot class distribution
sns.countplot(x=y, palette="coolwarm")
plt.title("Class Distribution (0 = Healthy, 1 = Parkinson's Disease)")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

```python
svm = SVC(kernel='rbf', C=1, gamma='scale')
svm.fit(X_train, y_train)
y_pred_svm = svm.predict(X_test)
print("\nSVM")
print("Accuracy:", accuracy_score(y_test, y_pred_svm))
print(classification_report(y_test, y_pred_svm))
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, accuracy_score

# Define MLP model
mlp = MLPClassifier(hidden_layer_sizes=(100, 50), max_iter=500, activation='relu', solver='adam', random_state=42)

# Train MLP
mlp.fit(X_train, y_train)

# Predict
y_pred_mlp = mlp.predict(X_test)

# Evaluate
mlp_accuracy = accuracy_score(y_test, y_pred_mlp)
print("MLP Classifier Accuracy:", mlp_accuracy)
print(classification_report(y_test, y_pred_mlp))
```

```python
X_lstm = X_scaled.reshape(-1, 1, X_scaled.shape[1])
X_train_lstm, X_test_lstm = X_lstm[:len(X_train)], X_lstm[len(X_train):]
model_lstm = Sequential([
    LSTM(64, input_shape=(1, X_scaled.shape[1]), return_sequences=False),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])
model_lstm.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model_lstm.fit(X_train_lstm, y_train, epochs=10, batch_size=32, verbose=0)
loss_lstm, acc_lstm = model_lstm.evaluate(X_test_lstm, y_test, verbose=0)
print("\nLSTM")
print("Accuracy:", acc_lstm)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt

#  1. Reshape the data for 1D CNN (samples, features, 1)
X_cnn = X_scaled.reshape(-1, X_scaled.shape[1], 1)
X_train_cnn, X_test_cnn = X_cnn[:len(X_train)], X_cnn[len(X_train):]

#  2. Define the CNN model architecture
model_cnn = Sequential([
    Conv1D(filters=32, kernel_size=3, activation='relu', input_shape=(X_scaled.shape[1], 1)),
    MaxPooling1D(pool_size=2),
    Flatten(),
    Dense(64, activation='relu'),
    Dropout(0.3),  # Prevent overfitting
    Dense(1, activation='sigmoid')  # Binary classification
])

#  3. Compile the model
model_cnn.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])

#  4. Train the model
history_cnn = model_cnn.fit(
    X_train_cnn, y_train,
    epochs=10,
    batch_size=32,
    validation_data=(X_test_cnn, y_test),
    verbose=1
)

#  5. Evaluate on test data
loss_cnn, acc_cnn = model_cnn.evaluate(X_test_cnn, y_test, verbose=0)
print("\n CNN Evaluation")
print("Test Accuracy:", round(acc_cnn * 100, 2), "%")

#  6. Plot training & validation accuracy/loss
plt.figure(figsize=(14, 5))

# Accuracy Plot
plt.subplot(1, 2, 1)
plt.plot(history_cnn.history['accuracy'], label='Train Accuracy', color='blue')
plt.plot(history_cnn.history['val_accuracy'], label='Validation Accuracy', color='green')
plt.title('CNN Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Loss Plot
plt.subplot(1, 2, 2)
plt.plot(history_cnn.history['loss'], label='Train Loss', color='red')
plt.plot(history_cnn.history['val_loss'], label='Validation Loss', color='orange')
plt.title('CNN Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
param_grid = {
    'n_estimators': [50, 100, 150],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}
grid_rf = GridSearchCV(RandomForestClassifier(random_state=42), param_grid, cv=3, scoring='accuracy')
grid_rf.fit(X_train, y_train)
print("\nRandom Forest Best Parameters:", grid_rf.best_params_)
print("Best Accuracy:", grid_rf.best_score_)
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Step: K-Nearest Neighbors (KNN)
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_test)

# Accuracy and classification report
knn_accuracy = accuracy_score(y_test, y_pred_knn)
print("K-Nearest Neighbors (KNN)")
print("Accuracy:", knn_accuracy)
print(classification_report(y_test, y_pred_knn))
```
